# Kalman Filter Example (1-D)

$\hat{X}_k = K_k Z_k + (1-K_k) \hat{X}_{k-1}$

Model  
$x_k = A x_{k-1} + B u_k + w_{k-1}$  
$z_k = H x_k + v_k$

Time Update (prediction)  
$\hat{x}^-_k = A \hat{x}_{k-1} + B u_k$  
$P^-_k = A P_{k-1} A^T + Q$

Measurement Updates (correction)  
$K_k = P^-_k H^T (H P^-_k H^T + R)^{-1}$  
$\hat{x}_k = \hat{x}^-_k + K_k(z_k - H \hat{x}^-_k)$  
$P_k = (I - K_k H)P^-_k$

```julia
using PyPlot

A = 1
B = 0
H = 1
Q = 0
R = 0.1

v = [0.39, 0.5, 0.48, 0.29, 0.25, 0.32, 0.34, 0.48, 0.41, 0.45]

x = 0.
P = 1.

r = zeros(length(v))

for k = 1:length(v)
    u = 0
    x_ = A * x + B * u
    P_ = A * P * A' + Q

    z = v[k]

    K = P_ * H' * inv(H * P_ * H' + R)
    x = x_ + K * (z - H * x_)
    P = (1 - K * H) * P_

    r[k] = x

    println("$k: $K $x $P")
end

plot(1:length(r), r);
ylim([0, 0.5])
grid(true)
```
